```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Extract Prouduct Title

def get_title(soup):
    try:
        title = soup.find("span",attrs={"id":'productTitle'}).text.strip()
    except AttributeError:
        title = ""
    
    return title


# Extract Product Price
def get_price(soup):
    try:
        price = new_soup.find("span",attrs={"class":"a-offscreen"}).text.strip()
    except AttributeError:
        price = ""

    return price

# Extract Product Rating
def get_rating(soup):
    try:
        rating = new_soup.find("span",attrs={"class":"a-icon-alt"}).text.strip()
    except AttributeError:
        rating = ""

    return rating[0:3]
    
```

```python
if __name__ == '__main__':

    URL = "https://www.amazon.com/s?k=samsung"
    USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36"

    # Header for request
    HEADERS = ({'User-Agent':USER_AGENT,'Accept-Language':'en-US , en;q=0.5'})

    # HTTP Request
    webpage = requests.get(URL,headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content,"html.parser")

    # Fetch links as List of all product.
    links = soup.find_all('a',attrs={'a-link-normal s-line-clamp-2 s-link-style a-text-normal'})

    # Store the links
    links_list = []

    # Loop for extracting links from tag
    for link in links:
        links_list.append(link.get('href'))

    # dictonary
    d = {"title":[],"price":[],"rating":[]}

    # loop for extracting product details from each link
    for link in links_list:
        new_webpage = requests.get("https://www.amazon.com" + link,headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content,"html.parser")

        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))

    
    amazon_df = pd.DataFrame.from_dict(d)
    # amazon_df['price'].replace('Page 1 of 1',np.nan,inplace=True)
    # amazon_df = amazon_df.dropna(subset=['price'])
    amazon_df = amazon_df[amazon_df['price'] != "Page 1 of 1"]
    amazon_df.to_csv("amaon_data.csv",header=True,index=False)
```
